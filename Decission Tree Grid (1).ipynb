```python
#importing the libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Reading the Dataset
dataset = pd.read_csv('insurance_pre.csv')
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male','smoker_yes']]
dependent=dataset['charges']
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
param_grid={'criterion':['squared_error', 'friedman_mse', 'absolute_error', 'poisson'],'max_features':[None,'sqrt','log2'],
             'splitter':['best','random']}
grid=GridSearchCV(DecisionTreeRegressor(),param_grid,refit=True,verbose=3,n_jobs=-1)
regressor=grid.fit(independent,dependent)
```

```python
y_pred=regressor.predict(independent)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(dependent,y_pred)
r_score
```

```python
re=grid.cv_results_
print("The R_score value for best parameter{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```

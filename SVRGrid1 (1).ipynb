```python
#importing the libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Reading the Dataset
dataset = pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
independent=dataset[['age', 'bmi', 'children','sex_male','smoker_yes']]
dependent=dataset['charges']
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid = {'kernel':['rbf','poly','sigmoid','linear'],
              'c':[10,100,1000,2000,3000],'gamma':['auto','scale']}
grid = GridSearchCV(SVR(),param_grid=param_grid,scoring='accuracy', refit = True, verbose = 3,n_jobs=-1)
regressor = grid.fit(independent,dependent)
```

```python
re=grid.cv_results_
print("The R_score value for best parameter ():".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(re)
```
